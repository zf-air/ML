```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('qiwen.csv')
```

```python
x=np.c_[df['height']]
y=np.c_[df['temperature']]
```

```python
plt.scatter(x,y)
plt.xlabel('GDP')
plt.ylabel('peri')
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
a=model.intercept_
b=model.coef_
print('a:',a,'b:',b)
```

```python
y_pre = model.predict(x)
```

```python
plt.plot(x,y_pre,color='red',linewidth=3,label='linear')
plt.scatter(x,y,color='blue',label='src')
plt.xlabel='gdp'
plt.ylabel='percentage'
plt.legend()
```
